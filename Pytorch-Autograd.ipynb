```python
import torch
```

```python
#create a Tensor

x = torch.tensor(3.0, requires_grad=True)
```

```python
y = x**2
```

```python
print(x)
print(y)
```

```python
# y.backward()
```

```python
x.grad
```

```python
z = torch.sin(y)
```

```python
print(x)
print(y)
print(z)
```

```python
z.backward()
```

```python
x.grad
```

```python
a = torch.tensor([1.0,2.0,3.0], requires_grad=True)
```

```python
a
```

```python
e
```

```python
e.backward()
```
